# 相关性检验

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 假设 df 是你的 DataFrame，columns_to_check 是你想要检查相关性的列名列表
columns_to_check = ['DLDZ_DQ200_SYSTEM_PI05.PV', 'DLDZ_AVS_SYSTEM_PI05.PV',
       'DLDZ_DQ200_LLJ01_FI01.PV', 'DLDZ_AVS_LLJ01_FI01.PV', '瞬时流量', '总压力',
       '星期', 'day', 'hour', 'minute', 'second'] 


# 选择指定的列
df_selected = df[columns_to_check]

# 计算相关系数矩阵
corr_matrix = df_selected.corr()

# 绘制热力图
plt.rcParams['font.sans-serif'] = ['SimSun']       # 设置字体为“宋体”，确保绘制图像不出错
plt.figure(figsize=(10, 8))  # 设置图表大小
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)  # annot=True 表示在单元格中显示数值
plt.title('相关性热力图')
plt.show()
```
