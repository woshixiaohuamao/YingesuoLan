```python
import pandas as pd

file_path = r'D:\Aa中工互联\工作安排\英格索兰\data\new历史数据.xlsx'
df = pd.read_excel(file_path)
```

# 获取瞬时流量

```python
df.columns
```

```python
df['时间'].head()
```

```python
df = df.sort_values(by='timestamp')
```

```python
# 对 'DQ200累积流量' 和 'AVS累积流量' 分别做一阶差分
df['DQ200差分值'] = df['DLDZ_DQ200_LLJ01_FQ01.PV'].diff()
df['AVS差分值'] = df['DLDZ_AVS_LLJ01_FQ01.PV'].diff()
```

```python
df.head()
```

```python
# 删除第一个NaN行（如果需要）
df.dropna(inplace=True)
df.head()
```

```python
df['瞬时流量'] = df['DQ200差分值'] + df['AVS差分值']
```

```python
df.head()
```

```python
df.to_excel('0瞬时流量.xlsx', index=False)
```
