```python
import pandas as pd

file_path = r'D:\Aa中工互联\工作安排\英格索兰\data\cleaned_data.xlsx'
df = pd.read_excel(file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------- 1. 准备数据 -----------
# 确保时间字段为datetime类型（重要！）
df['时间'] = pd.to_datetime(df['时间'])

# 按时间排序
df = df.sort_values('时间').reset_index(drop=True)

# ----------- 2. 创建画布 -----------
plt.figure(figsize=(16, 12))  # 调整画布大小
plt.rcParams['font.sans-serif'] = ['SimSun']       # 设置字体为“宋体”，确保绘制图像不出错
sns.set_style("whitegrid")     # 设置网格样式

# ----------- 3. 绘制四个子图 -----------
# DQ200系统压力
plt.subplot(2, 2, 1)  # 2行2列的第1个位置
sns.lineplot(x='时间', y='DQ200系统压力', data=df, color='royalblue', linewidth=1)
plt.title('DQ200系统压力趋势', fontsize=12, pad=10)
plt.xlabel('')  # 隐藏X轴标签（最下方子图统一显示）

# AVS系统压力
plt.subplot(2, 2, 2)
sns.lineplot(x='时间', y='AVS系统压力', data=df, color='crimson', linewidth=1)
plt.title('AVS系统压力趋势', fontsize=12, pad=10)
plt.xlabel('')

# DQ200累积流量
plt.subplot(2, 2, 3)
sns.lineplot(x='时间', y='DQ200累积流量', data=df, color='darkorange', linewidth=1)
plt.title('DQ200累积流量趋势', fontsize=12, pad=10)
plt.xlabel('时间', fontsize=10)  # 最下方子图显示X轴标签

# AVS累积流量
plt.subplot(2, 2, 4)
sns.lineplot(x='时间', y='AVS累积流量', data=df, color='forestgreen', linewidth=1)
plt.title('AVS累积流量趋势', fontsize=12, pad=10)
plt.xlabel('时间', fontsize=10)

# ----------- 4. 优化布局 -----------
plt.tight_layout(pad=3)  # 自动调整子图间距
plt.suptitle('系统参数趋势分析', y=1.02, fontsize=14, fontweight='bold')  # 添加总标题

# ----------- 5. 保存与显示 -----------
plt.savefig('trend_analysis.png', dpi=300, bbox_inches='tight')  # 保存高清图片
plt.show()
```
